```python
import requests

from bs4 import BeautifulSoup

import pandas as pd
```

```python
from datetime import timedelta, datetime, date, time

from io import StringIO
```

```python
headers = {
    'User-Agent':
    'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36'
}

session = requests.session()
```

```python
def getHistoricalStockData(stock_symbol, from_date, to_date) :
    
    session.get('https://www.nseindia.com', headers = headers)
    session.get(f'https://www.nseindia.com/get-quotes/equity?symbol={stock_symbol}', headers= headers) # To Save Cookies
    session.get(f'https://www.nseindia.com/api/historical/cm/equity?symbol={stock_symbol}',headers= headers)
    
    stock_url = f'https://www.nseindia.com/api/historical/cm/equity?symbol={stock_symbol}&series=[%22EQ%22]&from={from_date}&to={to_date}&csv=true'
    
    stock_data = session.get(url= stock_url, headers = headers)
    
    df = pd.read_csv(StringIO(stock_data.text[3:]))
    
    return df   
```

## Nifty Data

```python
def niftyHistoricalStockData(varient, from_date = ((datetime(datetime.today().year - 1, datetime.today().month, datetime.today().day) + timedelta(days=2)).strftime("%d-%m-%Y")), to_date =(datetime.today().strftime("%d-%m-%Y"))):
    varient = varient.upper()
    varient = varient.replace(' ','%20')
    varient = varient.replace('-','%20')
#     varient = varient.replace('&','%26')
    
    nifty_url = "https://www1.nseindia.com/products/dynaContent/equities/indices/historicalindices.jsp?indexType=" + varient +"&fromDate=" + from_date + "&toDate=" + to_date 
        
    nifty_data = session.get(url = nifty_url, headers = headers)
    
    soup = BeautifulSoup(nifty_data.text,'html5lib')
    
    df_nf = pd.read_csv(StringIO(soup.find('div',{'id':'csvContentDiv'}).contents[0].replace(':','\n')))
    
    return df_nf
```

```python
niftyHistoricalStockData('NIFTY 50')
```
